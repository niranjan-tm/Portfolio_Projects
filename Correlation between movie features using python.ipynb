```python
df =  pd.read_csv("E://acr//project 4//movies.csv")
```

```python
df.head()
```

```python
# Droping NaN values from data frame
df = df.dropna()
```

```python
df.head()
```

```python
# Finding correlations between movie features
```

```python
df.dtypes
```

```python
# droping released column
```

```python
df = df.drop(["released"],axis = 1)
```

```python
df.columns
```

```python
# sorting data frame by Gross earning
df = df.sort_values(by = ["gross"],inplace = False,ascending=False)
```

```python
df.head()
```

```python
df.budget = df.budget.astype("int64")
df.gross = df.gross.astype("int64")
```

```python
df.dtypes
```

```python
#find relation between Budget VS Gross Earning
plt.scatter(df.budget,df.gross);
plt.title("Movie Budget VS Gross Earning");
plt.xlabel("Movie Budget");
plt.ylabel("Gross earning");
```

```python
# As Budget of the movie is more chances are gross earning of the movie will increase 
```

```python
sns.regplot(data = df,x = df.budget,y = df.gross,scatter_kws={"color":"red"},line_kws={"color":"blue"});
plt.scatter(df.budget,df.gross);
plt.title("Movie Budget VS Gross Earning");
plt.xlabel("Movie Budget");
plt.ylabel("Gross earning");
```

```python
df.corr()
```

```python
sns.heatmap(data = df.corr(),annot=True);
```

```python
# Finding more correlation between movie features
```

```python
df1 = df
```

```python
for col in df1.columns:
    if df1[col].dtypes == "object":
        df1[col] = df1[col].astype("category")
        df1[col] = df1[col].cat.codes
```

```python
df1.head()
```

```python
sns.heatmap(data = df1.corr(),annot=True);
plt.title("Correlation between movie features");
```

```python
# From above Heat map it is clear that Gross earning of a movie is highly correlated to Budget and Votes of particular movie
```
